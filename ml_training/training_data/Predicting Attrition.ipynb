#Finding optimal cutoff

```python
FN_cost = 3.0
```

```python
FP_cost = 1.0
```

```python
def getCost(cutoff):
    y_df_v1['predicted_label'] = (y_df_v1['predicted'].map(lambda x: 1 if x > cutoff else 0))
    cm_v1 = confusion_matrix(y_df_v1.actual,y_df_v1.predicted_label,[1,0])
    return cm_v1[0][1] * FN_cost + cm_v1[1][0] * FP_cost
```

getCost(0.5)

```python
getCost(0.5)
```

```python
import numpy as np
```

```python
cutoff_list = np.arange(0.1,0.6,0.01)
```

```python
cutoff_list
```

```python
cost_list = [getCost(x) for x in cutoff_list]
```

```python
cost_list
```
